### implementation of Batch Gradient Descent(BGD)/ Normal Gradient descent

```python
import numpy as np
import matplotlib.pyplot as plt

A = np.array([[2, 0], [0, 3]])

# A unit square's corners
square = np.array([[0,0],[1,0],[1,1],[0,1],[0,0]])
transformed = square @ A.T  # apply A to each point

fig, ax = plt.subplots(1,2, figsize=(10,5))
ax[0].plot(square[:,0], square[:,1], 'b-o')
ax[0].set_title("Before: unit square")
ax[1].plot(transformed[:,0], transformed[:,1], 'r-o')
ax[1].set_title("After A: stretched 2x in x, 3x in y")
for a in ax:
    a.set_xlim(-1,4); a.set_ylim(-1,4); a.grid(True); a.set_aspect('equal')
plt.savefig('matrix_transform.png', dpi=100)
```

```python
# Rank 1 example: every column is a multiple of [1, 2]
A = np.array([[1, 2],
               [2, 4]])
print(np.linalg.matrix_rank(A))  # 1

# Every output Ax is some multiple of [1,2]
x1 = np.array([1, 0])
x2 = np.array([3, -1])
print(A @ x1)  # [1, 2]
print(A @ x2)  # [-3, -6]  -- still a multiple of [1,2]!
```

```python
v1 = np.array([1, 0, 0])
v2 = np.array([0, 1, 0])
v3 = np.array([2, 3, 0])  # = 2*v1 + 3*v2 -- redundant!

# v1, v2, v3 are NOT linearly independent (only 2 "real" directions among 3 vectors)
M = np.array([v1, v2, v3])
print(np.linalg.matrix_rank(M))  # 2, not 3
```

```python
# The vector [3, 1] in the standard basis
v = np.array([3, 1])

# A different basis: b1=[1,1], b2=[1,-1]
B = np.array([[1, 1],
               [1, -1]]).T  # columns are basis vectors

# Find coordinates of v in basis B: solve B @ c = v
c = np.linalg.solve(B, v)
print(c)  # [2, 1]  -- meaning v = 2*[1,1] + 1*[1,-1] = [3,1] ✓
```
